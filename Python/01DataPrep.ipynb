```python
# import required libraries

import json

import pandas as pd

import urllib.request as request
```

# Get Senator Info 

```python
# Read in the data from Json API

with request.urlopen('https://www.govtrack.us/api/v2/role?current=true&role_type=senator') as response:
        source = response.read()
        data = json.loads(source)
```

```python
# Create Empty lists

state_list = []

name_list = []

first_name_list = []

last_name_list = []

party_list = []

gender_list = []

startdate_list = []

enddate_list = []

phone_list = []

twitterid_list = []

youtubeid_list = []

website_list = []

link_list = []

senator_rank_list = []

bioguideid_list = []
```

```python
for i in range(100):
    
    # Extract the relevant information
    
    state = data['objects'][i]['state']
    
    name = data['objects'][i]['person']['sortname']
    
    firstname = data['objects'][i]['person']['firstname']
    
    lastname = data['objects'][i]['person']['lastname']
    
    gender = data['objects'][i]['person']['gender_label']
    
    party = data['objects'][i]['party']
    
    startdate = data['objects'][i]['startdate']
    
    enddate = data['objects'][i]['enddate']
    
    phone = data['objects'][i]['phone']
    
    bioguideid = data['objects'][i]['person']['bioguideid']
    
    twitterid = data['objects'][i]['person']['twitterid']
    
    youtubeid = data['objects'][i]['person']['youtubeid']
    
    website = data['objects'][i]['website']
    
    link = data['objects'][i]['person']['link']
    
    senatorrank = data['objects'][i]['senator_rank']
    
    
    # Appending to empty list
    
    state_list.append(state)
    
    party_list.append(party)
    
    name_list.append(name)
    
    first_name_list.append(firstname)
    
    last_name_list.append(lastname)
    
    gender_list.append(gender)
    
    startdate_list.append(startdate)
    
    enddate_list.append(enddate)
    
    phone_list.append(phone)
    
    twitterid_list.append(twitterid)
    
    youtubeid_list.append(youtubeid)
    
    website_list.append(website)
    
    link_list.append(link)
    
    senator_rank_list.append(senatorrank)
    
    bioguideid_list.append(bioguideid)
```

```python
data_combine = {'Name':name_list, 'FirstName':first_name_list, 'LastName':last_name_list, 'bioguide_id':bioguideid_list, 'State':state_list, 'Gender':gender_list, 'Party':party_list, 'StartDate':startdate_list, 'EndDate':enddate_list, 'Contact':phone_list, 'Rank':senator_rank_list, 'Twitter_ID':twitterid_list, 'Youtube_ID':youtubeid_list, 'Website':website_list, 'Link':link_list}
```

```python
# Converting lists to DataFrame

df = pd.DataFrame(data_combine)

df.head()
```

```python
data_combine2 = {'Name':name_list, 'OtherName':first_name_list + last_name_list, 'Rank':senator_rank_list, 'bioguideid':bioguideid_list,'Gender':gender_list, 'Party':party_list, 'EndDate':enddate_list, 'Contact':phone_list, 'Twitter_ID':twitterid_list, 'Youtube_ID':youtubeid_list, 'Website':website_list, 'Link':link_list}
```

```python
df['Full Name'] =  df['FirstName'] + ' ' +  df['LastName']

df['Term Date'] = df['StartDate'] + ' to ' + df['EndDate']

df.head()
```

```python
df_senator = df[['Full Name', 'Rank', 'bioguide_id', 'State', 'Gender', 'Party', 'Term Date', 'Website', 'Link']]
df_senator.head()
```
